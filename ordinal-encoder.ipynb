```python
data = ({'Name':["Parth", "Divit", "Vishrut", "Agrim", "Dhruv","Vishal","Vatsalya","Vaibhav"],
       'Education':['Bachelors','Masters','PHD','High School','Diploma','Masters','Bachelors','Bachelors'],
       })
df = pd.DataFrame(data)
print(df)
```

```python
columns = df.columns.tolist()
print(columns)
```

```python
def ordinal_encoder(df, column_to_encode):
    df_new = df.copy()
    education_mapping = {'High School': 0, 'Diploma':0, 'Bachelors': 1, 'Masters': 2, 'PHD': 3}
    df_new["Education"] = df_new["Education"].map(education_mapping)
    df_new.drop("Education", axis=1)
    print("Original Data: \n", df)
    print("\n")
    print("Ordinal Encoded Data: \n", df_new)
```

```python
ordinal_encoder(df, 'Education')
```
