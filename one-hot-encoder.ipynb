```python
#Test Code
data = {'Names': ['Vishrut', 'Agrim', 'Vishrut', 'Vishrut', 'Vaibhav', 'Agrim', 'Divit'],
        'Food_Order': ['Burger', 'Burger', 'Pizza', 'Salad', 'Pizza', 'Shake', 'Salad']}
df = pd.DataFrame(data)
print(df)
```

```python
def one_hot_encoder(dataframe, column_to_onehot):
    new_df = dataframe.copy()
    print("Entries in the columns \n", new_df[column_to_onehot])
    unique_vals = list(set(new_df[column_to_onehot]))
    print("Unique Values in", column_to_onehot, unique_vals)
    print("\n")

    for value in unique_vals:
        new_column = f"{column_to_onehot}_{value}"
        new_df[new_column] = np.where(new_df[column_to_onehot] == value, 1, 0) #condtion, value given if True, value given if False
            
    new_df = new_df.drop(column_to_onehot, axis=1)
    print("One Hot Encoded: \n", new_df)

    return 0
```

```python
print(df)
print("\n")
one_hot_encoder(df, 'Food_Order')
```
