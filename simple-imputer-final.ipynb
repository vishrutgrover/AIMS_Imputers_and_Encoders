```python
data = ({'Col1':[None, 10, 20, 30, None, 30, None, 20, 10],
         'Col2':[None, 5,None, 3, None, 2, 1, 2, 3],
         'Col3':[30, None, None, 10, 20, 30, 20, 10, None],
         'Col4':[None, 32, None, 60, None, 10, None, None, 20],
         'Col5':[None, 22, 10, 50, None, 20, None, 50, 20]
        })

df = pd.DataFrame(data)
print(df)
```

```python
data = df.convert_dtypes()
print(data)
```

```python
columns = data.columns.tolist()
columns
```

```python
def imputingmean(df):
    df_new = df.copy()
    for i in columns:
        df_new[i].fillna(int(df_new[i].mean()), inplace=True) #dataframe.fillna(dataframe.mean(), inplace = False) 
    return df_new
```

```python
def imputingmode(df):
    df_new = df.copy()
    for i in columns:
        df_new[i].fillna(int(max(df_new[i].mode())), inplace=True) #dataframe.fillna(dataframe.mean(), inplace = False) 
    return df_new
```

```python
def imputingmedian(df):
    df_new = df.copy()
    for i in columns:
        df_new[i].fillna(int(df_new[i].median()), inplace=True) #dataframe.fillna(dataframe.mean(), inplace = False) 
    return df_new
```

```python
def imputingconst(df):
    df_new = df.copy()
    for i in columns:
        df_new[i].fillna(int(0), inplace=True)
    return df_new
```

```python
def SimpleImputer(dataframe, mode):
    if (mode == "mean"):
        df = imputingmean(dataframe)
    elif (mode == "median"):
         df = imputingmedian(dataframe)
    elif (mode == "mode"):
        df = imputingmode(dataframe)
    else:
        df = imputingconst(dataframe)
        
    return df
```
